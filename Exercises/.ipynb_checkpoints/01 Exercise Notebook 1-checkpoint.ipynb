### 1b) Write a function `polyreg` to fit a polynomial of a given order to a dataset.
The inputs to the function are a data matrix of dimension $N \times 2$,  and $k \geq 0$, the order of the polynomial.   The function should compute the coefficients of the polynomial $\beta_0 + \beta_1 x +  \ldots  +\beta_k x^{k}$ via least squares regression, and should return the coefficient vector, the fit, and the vector of residuals.

If specified the degree $k$ is greater than or equal to $N$, then the function must fit an order $(N-1)$ polynomial and set the remaining coefficients to zero.

**NOTE**: You are *not* allowed to use the built-in function `np.polyfit`.

```python
def polyreg(data_matrix, k):
    x = data_matrix[:, 0]
    y = data_matrix[:, 1]
    N = len(x)

    # Determine the actual degree to fit
    k_fit = min(k, N - 1)

    # Construct the Vandermonde matrix A
    A = np.vander(x, k_fit + 1, increasing=True)

    # Solve for the coefficients beta using least squares
    # np.linalg.lstsq returns (coefficients, sum of squared residuals, rank, singular values)
    beta_actual, residuals_sum_of_squares, rank, singular_values = np.linalg.lstsq(A, y, rcond=None)

    # Calculate the fit
    y_fit = A @ beta_actual

    # Calculate the residuals
    residuals = y - y_fit

    # Adjust beta to match the requested k if k_fit < k
    if k_fit < k:
        beta = np.zeros(k + 1)
        beta[:len(beta_actual)] = beta_actual
    else:
        beta = beta_actual

    return beta, y_fit, residuals
```

Use the tests below to check the outputs of the function you have written:

```python
# Some tests to make sure your function is working correctly

xcol = np.arange(-1, 1.05, 0.1)
ycol = 2 - 7*xcol + 3*(xcol**2)  # We are generating data accoridng to y = 2 - 7x + 3x^2
test_matrix = np.transpose(np.vstack((xcol,ycol)))
test_matrix.shape

beta_test = polyreg(test_matrix, k=2)[0]
assert((np.round(beta_test[0], 3) == 2) and (np.round(beta_test[1], 3) == -7) and (np.round(beta_test[2], 3) == 3))
# We want to check that using the function with k=2 recovers the coefficients exactly

# Now check the zeroth order fit, i.e., the function gives the correct output with k=0
beta_test = polyreg(test_matrix, k=0)[0]
res_test = polyreg(test_matrix, k=0)[2] #the last output of the function gives the vector of residuals

assert(np.round(beta_test, 3) == 3.1)
assert(np.round(np.linalg.norm(res_test), 3) == 19.937)
```

### 1c) Use `polyreg` to fit polynomial models for the data in `xy_data` for $k=2,3,4$:

- Plot the fits for the three cases on the same plot together with the scatterplot of the data. The plots should be labelled and a legend included.
- Compute and print the SSE and $R^2$ coefficient for each of the three cases.
- Which of the three models you would choose? Briefly justify your choice.

```python
import matplotlib.pyplot as plt

x = xy_data[:, 0]
y = xy_data[:, 1]

# Initialize lists to store results
fits = {}
residuals = {}

# Polynomial orders to test
orders = [2, 3, 4]

plt.figure(figsize=(10, 7))
plt.scatter(x, y, label='Data points', alpha=0.7)

for k in orders:
    beta, y_fit, res = polyreg(xy_data, k)
    fits[k] = y_fit
    residuals[k] = res

    # Sort x values for plotting the fit line smoothly
    sort_indices = np.argsort(x)
    x_sorted = x[sort_indices]
    y_fit_sorted = y_fit[sort_indices]

    plt.plot(x_sorted, y_fit_sorted, label=f'Polynomial order {k} fit')

    # Calculate SSE
    sse = np.sum(res**2)

    # Calculate R-squared
    # Total sum of squares (SST)
    y_mean = np.mean(y)
    sst = np.sum((y - y_mean)**2)
    r_squared = 1 - (sse / sst)

    print(f"For k={k}:")
    print(f"  SSE: {sse:.3f}")
    print(f"  R-squared: {r_squared:.3f}")
    print("\n")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Fits for k=2, 3, 4')
plt.grid(True)
plt.legend()
plt.show()
```

### 1d)  For the model you have chosen in the previous part (either $k=2/3/4)$:

- Plot the residuals in a scatter plot.
- Plot a histogram of the residuals along with a Gaussian pdf with zero mean and the same standard deviation as the residuals.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Assuming the chosen model is k=4, based on previous output
k_chosen = 4
res_chosen = residuals[k_chosen]
x = xy_data[:, 0]

# Plot 1: Scatter plot of residuals
plt.figure(figsize=(10, 5))
plt.scatter(x, res_chosen, alpha=0.7)
plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label='Zero residual line')
plt.xlabel('x')
plt.ylabel('Residuals')
plt.title(f'Residuals Scatter Plot for k={k_chosen} model')
plt.grid(True)
plt.legend()
plt.show()

# Plot 2: Histogram of residuals with Gaussian PDF
plt.figure(figsize=(10, 5))

# Calculate mean and standard deviation of residuals
mu, std = np.mean(res_chosen), np.std(res_chosen)

# Plot histogram
plt.hist(res_chosen, bins=30, density=True, alpha=0.6, color='g', label='Residuals Histogram')

# Plot Gaussian PDF
xmin, xmax = plt.xlim()
x_pdf = np.linspace(xmin, xmax, 100)
p_pdf = norm.pdf(x_pdf, mu, std)
plt.plot(x_pdf, p_pdf, 'k', linewidth=2, label=f'Gaussian PDF (mean={mu:.2f}, std={std:.2f})')

plt.xlabel('Residuals')
plt.ylabel('Density')
plt.title(f'Histogram of Residuals and Gaussian PDF for k={k_chosen} model')
plt.grid(True)
plt.legend()
plt.show()
```
